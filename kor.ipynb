```python
import requests
from bs4 import BeautifulSoup


def CAS2Comp(CAS_id) : 
    url = "https://www.google.co.kr/search?q=" + CAS_id  + " 영화"
    source_code = requests.get(url)
    plain_text = source_code.text
    soup = BeautifulSoup(plain_text , "lxml")
    data =  soup.findAll("div",{"id":"resultStats"})

    # 아래의 과정은 제목만 뽑아 내기 위함
    data = str(data)
    data = data.split("<")[1]
    o = data.split()[-1].replace("\uac1c" , "")
    return o
```
